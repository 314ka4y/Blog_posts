```python
import matplotlib.pyplot as plt  #I will do these 2 lines only in my first example
%matplotlib inline

# First create a grid of plots. Now it will be one plot

fig, axs = plt.subplots(figsize = (4,2))  #figsize argument tells computer the size of our plot

# And make our first plot, simple line
axs.plot([3, 5, 7], [3,5,7])    # As you can see we used object oriented method. 
axs.set_xlabel("Wow, number")   # We defined X label
axs.set_ylabel("Boom!")         # And Y label too!
axs.set_title("My first graph") # Don't forget about the title
```

```python
# Method 1. The same results. A little bit different syntaxis. 
plt.plot([3, 5, 7], [3,5,7]);      # Draw a line
plt.xlabel("Number?!")             # We defined X label 
plt.ylabel("Double Boom!")         # And Y label too!
plt.title("My first graph, but with Matlab syntaxis") # Don't forget about the title
```

```python
# We can customize our plots in different aspects 
# linestyle
fig, axs = plt.subplots(figsize = (4,2)) 
axs.plot([3, 4 ,5], [3, 4, 5], linestyle='-')  # solid
axs.plot([3, 4 ,5], [4, 5, 6], linestyle='--') # dashed
axs.plot([3, 4 ,5], [5, 6, 7], linestyle='-.') # dashdot
axs.plot([3, 4 ,5], [6, 7, 8], linestyle=':')  # dotted
# Color
fig, axs = plt.subplots(figsize = (4,2)) 
axs.plot([3, 4 ,5], [3, 4, 5], color='pink')    # long version by name
axs.plot([3, 4 ,5], [4, 5, 6], color='b')       # short version
axs.plot([3, 4 ,5], [5, 6, 7], color='#AAFF22') # HEX code
axs.plot([3, 4 ,5], [6, 7, 8], color='0.5')     # Even grayscale

```

```python
fig, (left_draw, right_draw) = plt.subplots(ncols=2,figsize = (8,4)) #Created and unpacked objects
left_draw.plot([3, 4 ,5], [3, 4, 5], color='pink')   #left will be pink
left_draw.set_title("I am left")                     #left label
right_draw.plot([3, 4 ,5], [3, 4, 5], linestyle=':') #right will be dotted
right_draw.set_title("I am right");                  #right label
```

```python
# Lets have some fun!!!
# We will create grid with different functions graphs
import numpy as np

X_axis = np.linspace(-5,5,num=30)      # We create evenly spaced 30 points on X axis between -5 and 5
# X_axis = [-3,-2,-1,0,1,2,3]          # You can replace previous line of code with this one to make it easier 
Y_axis = []                            # Empty list for future Y values.

fig, axs = plt.subplots(nrows = 2, ncols=3,figsize = (12,8), tight_layout = True) # Generate grid 2x2

for i in list(range(0,6)):              # We need a loop to go over and make a magic
    row = i//3                          # // Floor division allows us to get N of row 
    column = i%3                        # % allows us to get N of column
#   print(f"Row {row} ")                # I recommend to uncomment this line when you debug your code fot the first time
#   print(f"Column {column} ")          # I recommend to uncomment this line when you debug your code fot the first time
    Y_axis = [x**(i+1) for x in X_axis] # Generate our Y values for each type of graph
    axs[row][column].plot(X_axis,Y_axis)# Make a plot
    axs[row][column].set_title(f"Visualisation of x^{i+1} function")  # Title for each plot
    axs[row][column].set_xlabel(f"X-label")                           # X label
    axs[row][column].set_ylabel(f"Y-label")                           # Y label

plt.savefig("Picture.png")              # Save our work
```

```python
import numpy as np
import seaborn as sns

X_axis = np.linspace(-5,5,num=30)      # We create evenly spaced 30 points on X axis between -5 and 5
# X_axis = [-3,-2,-1,0,1,2,3]          # You can replace previous line of code with this one to make it easier 
Y_axis = []                            # Empty list for future Y values.

fig, axs = plt.subplots(nrows = 2, ncols=3,figsize = (12,8), tight_layout = True) # Generate grid 2x2

for i in list(range(0,6)):              # We need a loop to go over and make a magic
    row = i//3                          # // Floor division allows us to get N of row 
    column = i%3                        # % allows us to get N of column
#   print(f"Row {row} ")                # I recommend to uncomment this line when you debug your code fot the first time
#   print(f"Column {column} ")          # I recommend to uncomment this line when you debug your code fot the first time
    Y_axis = [x**(i+1) for x in X_axis]
    sns.lineplot(x = X_axis, y = Y_axis, ax = axs[row][column])       # We made visuals using Seaborn, providing our axs object 
    axs[row][column].set_title(f"Visualisation of x^{i+1} function")  # Title for each plot
    axs[row][column].set_xlabel(f"X-label")                           # X label
    axs[row][column].set_ylabel(f"Y-label")                           # Y label

plt.savefig("SNS_Picture.png")          # Save our work    
   
```

```python
import matplotlib.pyplot as plt  #I will do these 2 lines only in my first example
%matplotlib inline
# First create a grid of plots
fig, ax = plt.subplots(figsize = (4,2)) 

# And make our first plot, simple line
ax.plot([3, 5, 7], [3,5,7])    # As you can see we used object oriented method. 
ax.set_xlabel("Wow, number")   # We defined X label 
ax.set_ylabel("Boom!")         # And Y label too!
ax.set_title("My first graph") # Don't forget about the title
```

```python
# Method 1. The same results. A little bit different syntaxis. 
plt.plot([3, 5, 7], [3,5,7]);
plt.xlabel("Number?!")   # We defined X label 
plt.ylabel("Double Boom!")         # And Y label too!
plt.title("My first graph, but with Matlab syntaxis") # Don't forget about the title
```

```python
x = [3, 4 ,5]
# We can customize our plots in different aspects 

# linestyle
fig, ax = plt.subplots(figsize = (4,2)) 
ax.plot([3, 4 ,5], [3, 4, 5], linestyle='-')  # solid
ax.plot([3, 4 ,5], [4, 5, 6], linestyle='--') # dashed
ax.plot([3, 4 ,5], [5, 6, 7], linestyle='-.') # dashdot
ax.plot([3, 4 ,5], [6, 7, 8], linestyle=':')  # dotted
```

```python
# Color
fig, ax = plt.subplots(figsize = (4,2)) 
ax.plot([3, 4 ,5], [3, 4, 5], color='pink')    # long version by name
ax.plot([3, 4 ,5], [4, 5, 6], color='b')       # short version
ax.plot([3, 4 ,5], [5, 6, 7], color='#AAFF22') # HEX code
ax.plot([3, 4 ,5], [6, 7, 8], color='0.5')     # Even grayscale
```

```python
fig, ax = plt.subplots(figsize = (4,2))         # Generate pair of objects
print(f"The type of our ax object: {type(ax)}") # Check the type of axis object
```

```python
fig, ax = plt.subplots(nrows = 2, ncols=2,figsize = (6,4))               # Generate pair of objects
print(f"The type of our ax object:{type(ax)}. The shape is {ax.shape},") # Check the type of axis object
```

```python
fig, ax = plt.subplots(nrows = 2, ncols=2,figsize = (6,4))               # Generate pair of objects
print(f"The type of our [0],[0] ax object:{type(ax[0][0])}")             # Check the type of one axis object from array
```
